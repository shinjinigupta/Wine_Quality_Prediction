# This is EDA for red-wine dataset.We have performed different visualization on the dataset to show the relationship among the independent and the dependent variables.

```python
#Import and read the red-wine file which is in CSV format using pandas.
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\varsh\OneDrive\Desktop\MSDA\SEM2\MACHINE_LEARNING\Project\datasets\winequality-red.csv",sep=',',header='infer')
df
```

```python
#Determine the relationship between the independent and the dependent variables via heat-map.
plt.subplots(figsize=(15, 10))
sns.heatmap(df.corr(), annot = True)
```

### Based on the analysis from the above heatmap, we can clearly determine that the attribute 'alcohol' and 'quality' has a correlation value of 0.48, which means there is a postive corelation between the two. In other words, as the value of alcohol increase, the value of quality also increases and vice versa. Whereas, there is a negative correlation seen between the variable 'volatile acidity' and 'quality' thats is -0.39, which means as one increases, the other decreases and vice versa.

```python
fig, ax1 = plt.subplots(3,4, figsize=(22,16))
k = 0
columns = list(df.columns)
for i in range(3):
    for j in range(4):
            sns.boxplot(df['quality'], df[columns[k]], ax = ax1[i][j])
            k += 1
plt.show()
```

### Based on the analysis from the above boxplot, we can see clearly see the outliers present for feature,also we can determine the mean value to see where most of the feature value falls within. 

```python
#Histogram for each features in the red-wine dataset.
plt.figure(figsize=(20,16))
for i, col in enumerate(list(df.columns.values)):
    plt.subplot(4,3,i+1)
    plt.hist(df[col], color='green')
    plt.grid()
    plt.title(col)
```

### The above histograms shows the different distribution of data for each feature.
