### This is the EDA for the white wine dataset. We have performed different visualizations on the dataset to show the relationship among the features and target variable. The same EDA will be performed for the red wine dataset as well.

```python
# Importing the libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Importing the dataset for white wine

df_white_wine= pd.read_csv('/Users/shinjinigupta/Desktop/SJSU/SPRING_2021/DATA_245/PROJECT/Datasets/winequality-white.csv')
df_white_wine
```

```python
#Analysis of relationship

plt.subplots(figsize=(15, 10))
sns.heatmap(df_white_wine.corr(), annot = True)
```

From the above heatmap we can see that the feature 'alcohol' has a strong correlation with our target variable 'quality', the value being 0.44.
'sulphates' is also strongly correlated with 'quality' with a value of 0.054.
But the features like 'citric acid' and 'residual sugar' are not at all strongly correlated with 'quality' having the values of -0.0092 and -0.098.

```python
# Boxplot analysis to show the outliers

fig, ax1 = plt.subplots(3,4, figsize=(22,16))
x = 0
columns = list(df_white_wine.columns)
for i in range(3):
    for j in range(4):
            sns.boxplot(df_white_wine['quality'], df_white_wine[columns[x]], ax = ax1[i][j])
            x += 1
plt.show()
```

Based on the analysis from the above boxplot, we can see clearly see the outliers present for feature, also we can determine the mean value to see where most of the feature value falls within.

```python
# Histograms for the features

plt.figure(figsize=(20,16))
for i, col in enumerate(list(df_white_wine.columns.values)):
    plt.subplot(4,3,i+1)
    plt.hist(df_white_wine[col], color='green',label='data')
    plt.grid()
    plt.title(col)
```

The above histograms of the different features and the target column shows the distribution of data for that column.
